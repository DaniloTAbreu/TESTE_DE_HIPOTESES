# <font color='blue'>Problema de Estatística: Diferença de satisfação de clientes com os sabores de iogurte</font>

# 1 - Definindo o problema de negócio

### Uma empresa de alimentos deseja comparar a satisfação dos clientes que experimentaram três diferentes sabores de um novo iogurte. Eles coletaram dados de satisfação dos clientes em uma escala de 1 a 10 (1 = muito insatisfeito, 10 = muito satisfeito) para cada um dos três sabores: Morango, Baunilha e Chocolate. A empresa quer saber se há diferenças significativas na satisfação dos clientes entre os diferentes sabores.

# 2 - Decisões

### O problema de negócio é estatístico. O teste de Kruskal-Wallis é um teste não paramétrico usado para comparar mais de dois grupos independentes. Ele é apropriado aqui porque:
### - Não assume que os dados seguem uma distribuição normal.
### - Pode ser usado para dados ordinais (como uma escala de satisfação de 1 a 10).

# 4 - Criando os Conjuntos de dados

```python
# Definindo os níveis de satisfação dos clientes para os três sabores
np.random.seed(42)
satisfacao_morango = np.random.randint(1, 11, 100)  # Escala de 1 a 10
satisfacao_baunilha = np.random.randint(1, 11, 100)  # Escala de 1 a 10
satisfacao_chocolate = np.random.randint(1, 11, 100)  # Escala de 1 a 10

# Criando um DataFrame para facilitar a visualização
data = {
    'Sabor': ['Morango']*100 + ['Baunilha']*100 + ['Chocolate']*100,
    'Satisfacao': np.concatenate([satisfacao_morango, satisfacao_baunilha, satisfacao_chocolate])
}
df = pd.DataFrame(data)
```

```python
# Visualizando as distribuições com boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sabor', y='Satisfacao', data=df, palette='Set2')
plt.title('Distribuição dos Níveis de Satisfação por Sabor de Iogurte')
plt.xlabel('Sabor')
plt.ylabel('Nível de Satisfação')
plt.show()
```

```python
# Visualizando as distribuições com gráficos de densidade
plt.figure(figsize=(12, 6))
sns.kdeplot(satisfacao_morango, label='Morango', color='red')
sns.kdeplot(satisfacao_baunilha, label='Baunilha', color='blue')
sns.kdeplot(satisfacao_chocolate, label='Chocolate', color='green')
plt.title('Distribuição de Densidade dos Níveis de Satisfação por Sabor de Iogurte')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

# 5 - Verificando as suposições, para aplicar o teste Kolmogorov-Smirnov:

## 5.1 Não assume que os dados seguem uma distribuição normal

### Não assumimos que os dados tem distribuição normal.

## 5.2 Dados com numeração ordinal

### Pode ser usado para dados ordinais (como uma escala de satisfação de 1 a 10).

# 6 - Teste de hipóteses

## 6.1 Teste Kruskal-Wallis

```python
# Realizando o Teste de Kruskal-Wallis
stat, p = kruskal(satisfacao_morango, satisfacao_baunilha, satisfacao_chocolate)
```

## 6.2 Análise do resultado

```python
print(f'Estatística Kruskal-Wallis: {stat}')
print(f'p-valor: {p}')
```

```python
# Interpretação dos resultados
if p < 0.05:
    print("Rejeitamos a hipótese nula. Pelo menos uma das distribuições dos níveis de satisfação é diferente.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidências suficientes para afirmar que as distribuições dos níveis de satisfação são diferentes.")
```

```python
### Conclusão: 
### Rejeitamos a hipótese nula. 
### Há evidências estatísticas de que pelo menos uma das distribuições dos níveis de satisfação é diferente.
```
